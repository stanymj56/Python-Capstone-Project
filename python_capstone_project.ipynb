```python
import pandas as pd
import numpy as np
df = pd.read_csv("/content/drive/MyDrive/ABC Company.xlsx - Sheet1.csv")
df
```

```python
#preprocessing
df.isnull().sum()
df['Height'] = np.random.randint(150,181,size=len(df))
print(df)

df.info()
df.describe()
```

```python
#Analysis Tasks
# Distribution of employees across each team
team_counts = df['Team'].value_counts()

# Percentage of total employees
team_percentage = (team_counts / len(df)) * 100

# Combining all teams and their respective counts and percentage
team_distribution = pd.DataFrame({
    'Count': team_counts,
    'Percentage': team_percentage
})

print(team_distribution)
##insight: The team New Oreleans Pelicans has the highest no.of members while the Minnesto Timberwolves and Orlando Magic have the fewest members.

```

```python
#Segregate employees based on their positions.
position_counts = df['Position'].value_counts()

#dataframing
position_distribution = pd.DataFrame({
    'Count': position_counts
})

print(position_distribution)
 ##insight: The most common position among the employees is SG, while the least common is C.
```

```python
#minimum age
min_age = df['Age'].min()
print(min_age)

#maximum age
max_age = df['Age'].max()
print(max_age)
```

```python
#The predominant age group among employees.
#identify the predominant age group
bins=[18,25,35,45]
labels=['18-25','26-35','36-45']

df['Age Group'] = pd.cut(df['Age'],bins=bins,labels=labels)

predominant_agegroup = df['Age Group'].value_counts().idxmax()
print(predominant_agegroup)
##insight:# Most of the employees are in the 26-35 age group(about 200+), while the fewest employees are in the range 36-45 age group with fewer than 50 people.Hence the predominant age group is 26-35 years old.
```

```python
#The team and position having the highest salary expenditure.
salary_grouped = df.groupby(['Team', 'Position'])['Salary'].sum()

highest_salary_expenditure = salary_grouped.idxmax()
print(highest_salary_expenditure)
##insight: # Most of the employees are in the 26-35 age group(about 200+), while the fewest employees are in the range 36-45 age group with fewer than 50 people.Hence the predominant age group is 26-35 years old.
```

```python
#Age and Salary correlation
import matplotlib.pyplot as plt

correlation=df['Age'].corr(df['Salary'])
print('Correlation between Age and Salary:',correlation)


plt.scatter(df['Age'],df['Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age vs Salary')
plt.show()
##insight: # There is a positive correlation approximately 0.21 between age and salary. However this doesnot guarantee that salary will increase with age.
```

```python
#Graphical Representation
# Visualization for the count and percentage of employees in each team
plt.figure(figsize=(12, 6))
plt.bar(team_distribution.index, team_distribution['Count'])
plt.xlabel('Team')
plt.ylabel('Count')
plt.title('Count of Employees per Team')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 12))
plt.pie(team_distribution['Count'], labels=team_distribution.index, autopct='%1.1f%%', startangle=120)
plt.title('Percentage Distribution of Employees per Team',size=15,color='Purple')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Visualization for the count of employees in each position
plt.figure(figsize=(10, 6))
plt.bar(position_distribution.index, position_distribution['Count'], color='blue',width = 0.4)
plt.xlabel('Position')
plt.ylabel('Count')
plt.title('Count of Employees per Position')
plt.tight_layout()
plt.show()
```

```python
# Visualization for the predominant age group
age_group_counts = df['Age Group'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(age_group_counts.index, age_group_counts.values, color='purple',width = 0.4)
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Distribution of Employees by Age Group')
plt.tight_layout()
plt.show()
```

```python
# Visualization for the highest salary expenditure by team and position
import matplotlib.pyplot as plt

salary_grouped = df.groupby(['Team', 'Position'])['Salary'].sum()

plt.figure(figsize=(25, 10))
salary_grouped.plot(kind='bar',width=0.7)
plt.xlabel('Team and Position')
plt.ylabel('Total Salary')
plt.title('Total Salary Expenditure by Team and Position')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
